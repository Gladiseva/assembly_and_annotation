```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data for plotting
data = {
    "Accession": ["Altai5", "Db1", "Kar1", "Rubezhnoe1", "St0", "TAIR10", "Had6"],
    "Number of Genes": [26905, 25167, 27273, 27312, 28018, 27416, 27600],
    "Percentage of Genes in Orthogroups": [99.2, 99.1, 99.5, 99.5, 99.2, 98.3, 97.3],
    "Percentage of Unassigned Genes": [0.8, 0.9, 0.5, 0.5, 0.8, 1.7, 2.7],
    "Percentage of Orthogroups Containing Species": [88.6, 86.9, 92.5, 87.8, 93.0, 92.1, 99.5],
    "Percentage of Genes in Species-Specific Orthogroups": [0.1, 0.1, 0.1, 0.2, 0.2, 0.7, 1.9]
}

df = pd.DataFrame(data)

# Filter the DataFrame to only include the selected accessions
selected_accessions = ["TAIR10", "Had6", "Db1", "Rubezhnoe1"]
df_filtered = df[df["Accession"].isin(selected_accessions)]

# Reorder the DataFrame to make "TAIR10" come first
df_filtered = df_filtered.set_index("Accession").reindex(["TAIR10", "Had6", "Db1", "Rubezhnoe1"]).reset_index()

# Define a more vibrant color palette for each accession
color_palette = {
    "TAIR10": "#6495ED",    # Light Blue
    "Had6": "#FF7F50",      # Light Coral
    "Db1": "#9370DB",       # Medium Purple
    "Rubezhnoe1": "#FFD700" # Soft Gold
}

# Create a figure with 4 subplots (1 row, 4 columns)
fig, axes = plt.subplots(1, 4, figsize=(20, 6))

# Plot the 'Number of Genes' barplot with updated colors
sns.barplot(data=df_filtered, x="Accession", y="Number of Genes", palette=color_palette, ax=axes[0])
axes[0].set_title("Number of Genes")
axes[0].set_ylabel("Number of Genes")
axes[0].set_xlabel("Accession")
axes[0].tick_params(axis="x", rotation=45)

# Plot the 'Percentage of Unassigned Genes' barplot with updated colors
sns.barplot(data=df_filtered, x="Accession", y="Percentage of Unassigned Genes", palette=color_palette, ax=axes[1])
axes[1].set_title("Percentage of Unassigned Genes")
axes[1].set_ylabel("Percentage (%)")
axes[1].set_xlabel("Accession")
axes[1].tick_params(axis="x", rotation=45)

# Plot the 'Percentage of Orthogroups Containing Species' barplot with updated colors
sns.barplot(data=df_filtered, x="Accession", y="Percentage of Orthogroups Containing Species", palette=color_palette, ax=axes[2])
axes[2].set_title("Percentage of Orthogroups Containing Species")
axes[2].set_ylabel("Percentage (%)")
axes[2].set_xlabel("Accession")
axes[2].tick_params(axis="x", rotation=45)

# Plot the 'Percentage of Genes in Species-Specific Orthogroups' barplot with updated colors
sns.barplot(data=df_filtered, x="Accession", y="Percentage of Genes in Species-Specific Orthogroups", palette=color_palette, ax=axes[3])
axes[3].set_title("Percentage of Genes in Species-Specific Orthogroups")
axes[3].set_ylabel("Percentage (%)")
axes[3].set_xlabel("Accession")
axes[3].tick_params(axis="x", rotation=45)

# Adjust layout for better readability
plt.tight_layout()
plt.show()
```
